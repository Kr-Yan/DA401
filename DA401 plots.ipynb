```python
#Import files
import seaborn as sns
import matplotlib.pyplot as plt
import os
import os.path
import io
import pandas as pd

path = os.path.join( "adult20.csv") #Get file path
g1= pd.read_csv(path) #Read the file
```

```python
#make a violin plot
g=sns.catplot(x="ORIENT_A", y="DEPFREQ_A",
            kind="violin", data=g1)  

#Reset the x-tick and y-tick values
g.set_xticklabels(['Gay/Lesbian','Straight','Bisexual','Something else',"I don't know the answer",'Refused','Not Ascertained',"Don't know"], rotation=40, ha="right")
g.set_yticklabels(["","",'Daily','Weekly','Monthly','A few times a year',"Never"], rotation=40, ha="right")
g.set(xlabel='Sexual Orientation', ylabel='Depression Frequncy')

```

```python
#make a violin plot
g=sns.catplot(x="ORIENT_A", y="ANXFREQ_A",
            kind="violin", data=g1)
#Reset the x-tick and y-tick values
g.set_xticklabels(['Gay/Lesbian','Straight','Bisexual','Something else',"I don't know the answer",'Refused','Not Ascertained',"Don't know"], rotation=40, ha="right")
g.set_yticklabels(["","",'Daily','Weekly','Monthly','A few times a year',"Never"], rotation=40, ha="right")
g.set(xlabel='Sexual Orientation', ylabel='Anxiety Frequncy')

      
```

```python
#Make a line plot
a=sns.lineplot(data=g1, x="AGEP_A", y="DEPFREQ_A")
a.set(xlabel='Age', ylabel='Depression Frequncy')
#Reset the y-axis range
plt.ylim(3,5)
a.set_yticklabels(['Monthly','','','','A few times a year','','','',"Never"], rotation=40, ha="right")
```

```python
#Make a line plot
a=sns.lineplot(data=g1, x="AGEP_A", y="ANXFREQ_A")
a.set(xlabel='Age', ylabel='Anxiety Frequncy')
#Reset the y-axis range
plt.ylim(2,5)
a.set_yticklabels(['Weekly','','Monthly','','A few times a year','',"Never"], rotation=40, ha="right")
```

```python
#Make a point plot
b=sns.catplot(x="SLPFLL_A", y="DEPFREQ_A",markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=g1)
#Reset the x-tick and y-tick
b.set_xticklabels(["Never","Some days", "Most days",
 "Every day", "Refused","Not Ascertained", "Don't Know"],rotation=40, ha="right")
b.set_yticklabels(["","Monthly","","","","A few times a year","","","","Never"], rotation=40)
#Reset x, y-axis name and range
b.set(xlabel='Frequency of Trouble Falling Asleep', ylabel='Depression Frequncy')
plt.ylim(2.75,5)
```

```python
#Make a point plot
b=sns.catplot(x="SLPFLL_A", y="ANXFREQ_A",markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=g1)

#Reset the x-tick and y-tick
b.set_xticklabels(["Never","Some days", "Most days",
 "Every day", "Refused","Not Ascertained", "Don't Know"],rotation=40, ha="right")
b.set_yticklabels(["","","Monthly",'',"A few times a year","",""], rotation=40)

#Reset x, y-axis name and range
b.set(xlabel='Frequency of Trouble Falling Asleep', ylabel='Anxiety Frequncy')
plt.ylim(2.25,4.75)
```

```python
#Make a boxplot
t = sns.boxplot(x="ASEV_A", y="ANXFREQ_A", data=g1)

#Reset x, y-axis
t.set_xticklabels(["Yes", "No", "Refused","Don't Know"],rotation=40, ha="right")
t.set_yticklabels(["","Daily","Weekly","Monthly","A few times a year","Never"], rotation=40)
t.set(xlabel='Ever had asthma', ylabel='Anxiety Frequncy')

plt.ylim(0.5,5.5)
```
